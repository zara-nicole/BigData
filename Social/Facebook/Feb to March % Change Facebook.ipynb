```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
import glob
import re
import functions_social as fs
```

```python
%store -r my_data
%store -r fb_data
```

```python
fb_year = fb_data[['2017-03','2018-03','2017-02','2018-02']]
fb_year['perc_change3'] = (fb_data['2018-03']-fb_data['2017-03'])*100
fb_year['perc_change2'] = (fb_data['2018-02']-fb_data['2017-02'])*100
fb_year_sorted = fb_year.sort_values('perc_change3',ascending=False).reset_index().rename(index=str,columns={'index':'website'})
fb_year_sorted
```

```python
from bokeh.layouts import column
from bokeh.io import show, output_file, output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.plotting import figure
from bokeh.palettes import Spectral5
from bokeh.transform import factor_cmap
output_notebook()
group = fb_year_sorted
tools = 'pan,wheel_zoom,xbox_select,reset'
source = ColumnDataSource(group)
output_file("x.html")
p = figure(plot_width=4000, plot_height=500, title="% Change in Facebook Share (2017 to 2018)",
           x_range=fb_year_sorted['website'], toolbar_location=None, tools=tools,
           active_drag="xbox_select")

p.vbar(x='website', top='perc_change3', width=1, source=source,
       line_color="white", fill_color='blue' )

p.y_range.start = -80
p.x_range.range_padding = 0.01
p.xgrid.grid_line_color = None
p.xaxis.axis_label = "Website"
p.xaxis.major_label_orientation = 1.2
p.outline_line_color = None
p.title.text_font_size = '13pt'
p.title.text_color = 'black'
#p.title.align = 'center'
p.add_tools(HoverTool(tooltips=[("% Change", "@perc_change3"), ("Website", "@website")]))

q = figure(plot_width=4000, plot_height=500, title="% Change in Facebook Share (2017-02 to 2018-02)",
           x_range=fb_year_sorted['website'], toolbar_location=None, tools=tools,
          active_drag="xbox_select")

q.vbar(x='website', top='perc_change2', width=1, source=source,
       line_color="white", fill_color='blue' )

q.y_range.start = -80
q.x_range.range_padding = 0.01
q.xgrid.grid_line_color = None
q.xaxis.axis_label = "Website"
q.xaxis.major_label_orientation = 1.2
q.outline_line_color = None
q.x_range = p.x_range
q.title.text_font_size = '13pt'
q.title.text_color = 'black'

q.add_tools(HoverTool(tooltips=[("% Change", "@perc_change2"), ("Website", "@website")]))
show(column(p,q))
```
