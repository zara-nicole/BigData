```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
```

```python
%store -r df_fb
%store -r df_twi
import pandas as pd
```

```python
class FixedOrderFormatter(ScalarFormatter):
    """Formats axis ticks using scientific notation with a constant order of 
    magnitude"""
    def __init__(self, order_of_mag=0, useOffset=True, useMathText=False):
        self._order_of_mag = order_of_mag
        ScalarFormatter.__init__(self, useOffset=useOffset, 
                                 useMathText=useMathText)
    def _set_orderOfMagnitude(self, range):
        """Over-riding this to avoid having orderOfMagnitude reset elsewhere"""
        self.orderOfMagnitude = self._order_of_mag
```

```python
def weight_avg(df,col,name1):
    #input: dataframe, usernames to compare (printed in describe_data)
    #example: compare_likes(data,'jackinthebox','burgerking')
    sns.set(style="ticks")
    fig, ax1 = plt.subplots(figsize=(12,8))
    fig.suptitle('Weighted Average of FB Likes and Twitter Followers',y=1)

    color = 'tab:red'
    ax1.set_xlabel('Time (months)',labelpad=15,fontsize=15)
    ax1.set_ylabel(name1, color=color,labelpad=15,fontsize=15)
    df_1 = df[df['Username']==name1]
    ax1.plot(df_1['As Of Date'],df_1[col], color=color)
    ax1.tick_params(axis='y', labelcolor=color,labelsize=11)
    ax1.tick_params(axis='x', labelsize=11)
    ax1.ticklabel_format(axis='y',style='sci')
    ax1.yaxis.set_major_formatter(FixedOrderFormatter(6))
    
    fig.tight_layout()  # otherwise the right y-label is slightly clipped
    plt.show()
```

```python
tw_clean = df_twi[['Ticker Symbol','As Of Date','Screen Name','Followers']]
tw_clean = tw_clean[tw_clean['Screen Name'].isin(['gucci','YSL','BALENCIAGA','Prada','Dior','LouisVuitton','marcjacobs'])]
```

```python
tw_clean.columns=['Ticker Symbol','As Of Date','Username','Followers']
tw_clean['Username'] = tw_clean['Username'].str.replace('BALENCIAGA','Balenciaga')
tw_clean['Username'] = tw_clean['Username'].str.replace('gucci','Gucci')
```

```python
tw_clean.head()
```

```python
tw_gucci = tw_clean[tw_clean['Username']=='Gucci']
fb_gucci = fb_clean[fb_clean['Username']=='Gucci']
df_gucci = pd.merge(fb_gucci,tw_gucci,on=['Username','As Of Date'],how='inner')
df_gucci['Weighted Likes'] = pd.Series(df_gucci['Likes']/(df_gucci['Likes']+df_gucci['Followers']),index=df_gucci.index)
df_gucci['Weighted Followers'] = pd.Series(df_gucci['Followers']/(df_gucci['Likes']+df_gucci['Followers']),index=df_gucci.index)
df_gucci['Weighted Average'] = pd.Series((df_gucci['Weighted Likes']*df_gucci['Likes']+df_gucci['Followers']*df_gucci['Weighted Followers']),index=df_gucci.index)
```

```python
df_gucci['Shift'] = pd.Series(df_gucci['Weighted Average'].shift(1),index=df_gucci.index)
df_gucci['Diff'] = pd.Series((df_gucci['Shift']-df_gucci['Weighted Average']),index=df_gucci.index)
df_gucci.std()
```

```python
weight_avg(df_gucci,'Weighted Average','Gucci')
```
